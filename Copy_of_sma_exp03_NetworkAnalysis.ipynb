```python
import networkx as nx
```

```python
import matplotlib.pyplot as plt
```

```python
G=nx.Graph()
```

```python
G.add_nodes_from(['a','b','c','d'])
```

```python
G.nodes()
```

```python
G.add_edge('a','b')
G.add_edge('a','c')
G.add_edge('b','d')
```

```python
G.edges()
```

```python
nx.draw(G)
plt.show()
```

```python
nx.degree(G)
```

```python
nx.degree_centrality(G)
```

```python
nx.shortest_path(G,'a','d')
```

```python
nx.betweenness_centrality(G)
```

```python
nx.eigenvector_centrality(G)
```

```python
G=nx.gnp_random_graph(10,0.5)
nx.draw(G)
```

```python
nx.draw(G,with_labels=1)
plt.show()
```

```python
nx.degree(G)
```

```python
nx.degree_centrality(G)
```

```python
nx.betweenness_centrality(G)
```

```python
nx.eigenvector_centrality(G)
```

```python
m_influential=nx.degree_centrality(G)
```

```python
for w in sorted(m_influential,key=m_influential.get,reverse=True):
  print(w,m_influential[w])
```

```python
M=nx.Graph()
```

```python
M.add_nodes_from(['a','b','c','d','e','f','g','h'])
```

```python
M.nodes()
```

```python
M.add_edge('a','b')
M.add_edge('b','c')
M.add_edge('b','d')
M.add_edge('c','e')
M.add_edge('d','e')
M.add_edge('e','h')
M.add_edge('e','f')
M.add_edge('e','g')
M.add_edge('d','g')
M.add_edge('g','h')
```

```python
nx.draw(M)
plt.show()
```

```python
nx.draw(M,with_labels='a')
plt.show()
```

```python
nx.degree(M)
```

```python
nx.degree_centrality(M)
```

```python
nx.betweenness_centrality(M)
```

```python
nx.eigenvector_centrality(M)
```

```python
m_influential=nx.betweenness_centrality(M)
```

```python
for x in sorted(m_influential,key=m_influential.get,reverse=True):
  print(x,m_influential[x])
```
